```python
!pip3 install openpyxl
import pandas as pd

# Load the dataset
file_path = "/Users/himanshumaurya/Desktop/Dashboard/EMDAT Custom Request Apr 2025.xlsx"  # Replace with the actual file path
df = pd.read_excel(file_path, engine="openpyxl")

# Display column names and first few rows
print(df.head())
print(df.columns)
disaster_counts = df["Disaster Type"].value_counts()
print(disaster_counts)
```

```python
# Select the top 9 disaster types
top_disasters = disaster_counts.nlargest(9).index.tolist()
print("Selected Disasters:", top_disasters)

# Filter dataset for selected disasters
df_filtered = df[df["Disaster Type"].isin(top_disasters)]

# Select relevant columns
columns_needed = ["Year", "Disaster Type", "Country", "Latitude", "Longitude", 
                  "Total Deaths", "Total Affected", "Economic Damage"]
df_filtered = df_filtered[columns_needed]

# Drop rows with missing geolocation data
df_filtered = df_filtered.dropna(subset=["Latitude", "Longitude"])

# Save filtered data to CSV
filtered_csv_path = "top_9_disasters_filtered.csv"
df_filtered.to_csv(filtered_csv_path, index=False)

print(f"Filtered data saved successfully as {filtered_csv_path}")
```
